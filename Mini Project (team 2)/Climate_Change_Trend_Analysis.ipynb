```python
# 🧹 Data Cleaning

#The dataset contains many columns. For this project, only the most relevant features are selected to simplify the analysis.
```

```python
# Select Important Columns

climate = df[['country',
              'year',
              'co2',
              'co2_per_capita',
              'population',
              'gdp']]

climate.head()
```

```python
climate.isnull().sum()#to check missing values
```

```python
# Remove missing values

climate = climate.dropna()

# Remove duplicate rows

climate = climate.drop_duplicates()

print("Dataset Shape After Cleaning:", climate.shape)
```
